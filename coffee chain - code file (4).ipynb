```python
def exp_data_analysis():
    print("This function will do the following data exploration functions as per your choice: head/tail/sample/info/describe/shape- ")
    functions = ['head', 'tail', 'sample', 'info', 'describe', 'shape', 'columns', 'ndim']
    for i in functions:
        
        op = input("What exploratory data analysis function would you like to perform?: ")
        if op == 'head':
            
            val = int(input("How many values would you like?"))
            print(df.head(val))
        elif op == 'tail':
            val = int(input("How many values would you like?"))
            print(df.tail(val))
        elif op == 'sample':
            val = int(input("How many values would you like?"))
            print(df.sample(val))
        elif op  == 'describe':
            print(df.describe())
        elif op == 'shape':
            print(df.shape)
        elif op  == 'info':
            print(df.info())
        elif op == 'columns':
            print(df.columns)
        
        else:
            print("please choose a valid action from the list.")
            break 
            
exp_data_analysis()
```

```python
def handle_null_val():
    global df
    print("Do you wish to delete the rows with null values or fill them?")
    g = input("enter fill or delete:")

    if g == "delete":
        print(df.dropna())
    elif g == "fill":
        print("You can either fill the null values with the mean, backward fill or forward fill")
        g = input("enter either mean, backward fill or forward fill: ")

        if g == "mean":
            numeric_columns = df.select_dtypes(include='number') 
            df[numeric_columns.columns] = df[numeric_columns.columns].fillna(numeric_columns.mean()) 
            print("Null values have been mean-filled.")
            print(df)
        elif g == "backward fill":
            df.fillna(method='bfill', inplace=True)
            print("Null values have been backward filled.")
            
            print(df)
        elif g == "forward fill":
            df.fillna(method='ffill', inplace=True)
            print("Null values have been forward filled.")
            print(df)

handle_null_val()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='Product', y='Profit', data=df, color='tan', errorbar=None)
plt.title('Product vs Profit')
plt.xlabel('Product')
plt.ylabel('Profit')
plt.xticks(rotation=45, ha='right')

plt.show()


f_grouped = df.groupby('Product').agg({'Marketing': 'sum', 'Profit': 'sum'}).reset_index()
sns.barplot(x='Product', y='Profit', hue='Marketing',palette='light:brown', data=f_grouped)
plt.title('Profit and Marketing by Product')
plt.xlabel('Product')
plt.ylabel('Marketing Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
market_counts = df['Market'].value_counts()
colors = ['tan', 'wheat', 'beige', '#DEB887', '#D4A5A5']  
market_counts.plot.pie(autopct='%1.1f%%', startangle=90, colors=colors)
plt.title('Distribution of Market')
plt.ylabel('')
plt.show()
```

```python
market_product_crosstab = pd.crosstab(df['Market'], df['Product Type'])
light_brownish_colors = ['#D6C4A1', '#B89A6A', '#C4A785', '#D8B28D']
for market in market_product_crosstab.index:
    plt.figure(figsize=(8, 6))
    market_data = market_product_crosstab.loc[market]
    market_data.plot.pie(autopct='%1.1f%%', startangle=90, colors = light_brownish_colors)
    plt.title(f'Distribution of Product Type in {market}')
    plt.ylabel('') 
    plt.show()
```

```python
df_grouped = df.groupby('Market').agg({'Total Expenses': 'sum', 'Profit': 'sum'}).reset_index()
color_palette = {'Total Expenses': '#F2D1B3', 'Profit': '#B4B8B1'}
sns.barplot(x='Market', y='value', palette=color_palette, hue='variable', data=pd.melt(df_grouped, id_vars=['Market'], var_name='variable', value_name='value'))
plt.title('Total Expenses and Profit by Market')
plt.xlabel('Market')
plt.ylabel('Amount')
plt.xticks(rotation=45)
plt.show()
```

```python
df['Ddate'] = pd.to_datetime(df['Ddate'], format='mixed')
df['Month'] = df['Ddate'].dt.month
monthly_sales = df.groupby('Month')['Coffee Sales'].sum()
plt.figure(figsize=(10, 6))
plt.plot(monthly_sales, marker='o', color='black')
plt.xlabel('Month')
plt.ylabel('Total Coffee Sales')
plt.title('Monthly Coffee Sales Trend')
plt.xticks(range(1, 13)) 
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df['Ddate'] = pd.to_datetime(df['Ddate'], format='mixed')
df['Inventory'] = pd.to_numeric(df['Inventory'], errors='coerce')
df['Year'] = df['Ddate'].dt.year
yearly_data = df.groupby('Year')[['Coffee Sales', 'Inventory']].sum()
yearly_data.plot(kind='bar', figsize=(10, 6), color=['tan', 'maroon'])
plt.xlabel('Year')
plt.ylabel('Amount in 000s')
plt.title('Yearly Comparison of Sales vs. Inventory')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
aggregated_data = df.groupby('Market')[['Profit', 'Margin', 'Coffee Sales']].sum()

fig, axes = plt.subplots(1, 2, figsize=(18, 8))

sns.heatmap(aggregated_data, annot=True, cmap='YlGnBu', fmt='g', cbar=True, linewidths=0.5, ax=axes[0])
axes[0].set_title('Heatmap of Key Metrics by Market')
axes[0].set_xlabel('Market')
axes[0].set_ylabel('Metrics')

aggregated_data_reset = aggregated_data.reset_index()
df_melted = aggregated_data_reset.melt(id_vars=['Market'], value_vars=['Profit', 'Margin', 'Coffee Sales'],
                                       var_name='Metric', value_name='Value')
sns.barplot(x='Market', y='Value', hue='Metric', data=df_melted, palette='Set2', ax=axes[1])
axes[1].set_title('Comparison of Key Metrics by Market')
axes[1].set_xlabel('Market')
axes[1].set_ylabel('Value')

plt.tight_layout()
plt.show()
```

```python
colors = plt.cm.tab10(np.linspace(0, 1, len(df['Product'].unique())))
plt.figure(figsize=(12, 8))  
for i, product in enumerate(df['Product'].unique()):
    product_data = df[df['Product'] == product]
    plt.scatter(product_data['Marketing'], product_data['Profit'], color=colors[i], label=product)

plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left', borderaxespad=0)  
plt.xlabel('Marketing Expenditure')
plt.ylabel('Profit')
plt.title('Profit vs. Marketing by Product')
plt.show()
```

```python
grouped_df = df.groupby('Market')
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(13, 10))
axes = axes.flatten()
for i, (market, group_df) in enumerate(grouped_df):
    ax = axes[i]
    group_df.set_index('Budget Sales', inplace=True)
    group_df['Coffee Sales'].plot(ax=ax)
    ax.set_title(f"{market} Market")
    ax.set_xlabel("Budget sales")
    ax.set_ylabel("Coffee Sales")
plt.tight_layout()
plt.show()
```
